```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import numpy as np
```

```python
# Load dataset
try:
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
    print(" Dataset loaded successfully.")
except Exception as e:
    print(" Error loading dataset:", e)
```

```python
# Display first few rows
print(df.head())
```

```python
# Check for missing values
print("\nMissing values:\n", df.isnull().sum())
```

```python
# Data types
print("\nData types:\n", df.dtypes)
```

```python
# Basic statistics
print("\nBasic Statistics:\n", df.describe())
```

```python
# Group by species
grouped = df.groupby("species").mean()
print("\nMean values per species:\n", grouped)
```

```python
# Observations
print("\nObservations:")
print("1. Setosa species tends to have smaller measurements across all features.")
print("2. Virginica species generally has the largest measurements.")
print("3. Versicolor falls between Setosa and Virginica in most measurements.")
```

```python
# === Data Visualization ===

# Line chart (fake trend using index)
plt.figure(figsize=(8, 4))
plt.plot(df.index, df["sepal length (cm)"], label="Sepal Length")
plt.title("Sepal Length Over Observations")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Bar chart: Average petal length per species
plt.figure(figsize=(6, 4))
grouped["petal length (cm)"].plot(kind='bar', color='skyblue')
plt.title("Average Petal Length by Species")
plt.ylabel("Petal Length (cm)")
plt.xlabel("Species")
plt.tight_layout()
plt.show()
```

```python
# Histogram: Distribution of sepal width
plt.figure(figsize=(6, 4))
plt.hist(df["sepal width (cm)"], bins=15, color='orange', edgecolor='black')
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Sepal length vs. petal length
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x="sepal length (cm)", y="petal length (cm)", hue="species")
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()
plt.tight_layout()
plt.show()
```
